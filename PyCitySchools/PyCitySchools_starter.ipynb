```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Calculate and print the total number of schools
count_schools = school_data["School ID"].count()
count_schools
```

```python
# Calculate and print the total number of students
count_students = student_data["Student ID"].count()
count_students
```

```python
# Calculate and print the total budget
sum_district_budget = school_data["budget"].sum()
sum_district_budget
```

```python
# Calculate and print the average math score
avg_student_math = student_data["math_score"].mean()
avg_student_math
```

```python
# Calculate and print the average reading score
avg_student_reading = student_data["reading_score"].mean()
avg_student_reading
```

```python
# Initialize the passing grade
passing_score = 70
```

```python
# Add a Boolean to the combined set to determine whether each student had a passing score for math and print the header
school_data_complete["passing_math_yn"] = school_data_complete.math_score >= passing_score
school_data_complete.head()
```

```python
# Calculate and print the percentage of students who have a passing math score
pc_student_passing_math = school_data_complete["passing_math_yn"].mean() * 100
pc_student_passing_math
```

```python
# Add a Boolean to the combined set to determine whether each student had a passing score for reading and print the header
school_data_complete["passing_reading_yn"] = school_data_complete.reading_score >= passing_score
school_data_complete.head()
```

```python
# Calculate and print the percentage of students who have a passing math score
pc_student_passing_reading = school_data_complete["passing_reading_yn"].mean() * 100
pc_student_passing_reading
```

```python
# Calculate and print the "overall passing rate"

#     Note: it's strange to me that we are framing the average overall score as "overall passing rate" ...
# ... which I interpret to mean we should calculate each student's average score ...
# ... and compare that average to the pass/fail line of 70 ...
# ... and then pass that in as another Boolean column in our dataframe ...
# ... which could give us a sense of what percentage of students are passing in an OVERALL sense.
# 
#     Something like:
#     school_data_complete["passing_overall_yn"] =
#     (school_data_complete.reading_score + school_data_complete.math_score) / 2 >= passing_score

#     In reviewing the provided answer key, however, it looks like "overall passing rate" ...
# ... expects the average of the math and reading AVERAGES when compiling the district summary ...
# ... BUT
# ... expects the average of the math and reading PASSING RATES when compiling the other summaries given.
# 
#     Perhaps this inconsistency is intentional, but I feel providing unclear/inconsistent instructions ...
# ... might penalize students, especially those of us who prefer to work without referencing the answer key.
# 
#     Out of a desire to make it easier for our over-worked TAs to review my submission ...
# ... I am submitting code that I feel will best match the answer key ...
# ... rather than what I believe would be the best (or at least internally consistent) analysis ...
# ... and hope that future requirements can be reviewed for clarity/consistency prior to assignment.

avg_student_overall = (avg_student_math + avg_student_reading) / 2
avg_student_overall
```
